```python
# Cargamos la base de datos de Junio
df0=pd.read_csv('BD_22jun.txt',sep='\s+')
# Le damos la estructura de datos
dfJ=pd.DataFrame(df0)
```

```python
# Cargamos la base de datos de Diciembre
df1=pd.read_csv('BD_22dic.txt',sep='\s+')
# Le damos la estructura de datos
dfD=pd.DataFrame(df1)
```

```python
plt.title('Presión vs. Altura 22 de Junio')
plt.ylabel('Altura (m)')
plt.xlabel('Presión (hPa)')
plt.grid(True)

x=dfJ["PRES"]
y=dfJ["HGHT"]

plt.plot(x,y)
plt.show()
```

```python
plt.title('Temperatura y Temp. de Rocío vs. Altura 22 de Junio')
plt.ylabel('Altura (m)')
plt.xlabel('Temp\TempR (C)')
plt.grid(True)

x=dfJ[["TEMP","DWPT"]]
y=dfJ["HGHT"]
plt.legend(loc='best')
plt.plot(x,y)
plt.show()
```
