```python
################
## Ejemplo 3.3 #
################


def vectorfield(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x1,y1,x2,y2]
        t :  time
        p :  vector of the parameters:
                  p = [m1,m2,k1,k2,L1,L2,b1,b2]
    """
    x1, y1, x2, y2 = w
    m1, m2, k1, k2, L1, L2, b1, b2, mu1, mu2 = p

    # Create f = (x1',y1',x2',y2'):
    f = [y1,
         (-b1 * y1 - k1 * (x1) + k2 * (x2 - x1 )+ mu1*((x1)**3)+mu2*((x1-x2)**3)) / m1,
         y2,
         (-b2 * y2 - k2 * (x2 - x1)+mu2*((x2-x1)**3)) / m2]
    return f
```

```python
# Parameter values
# Masses:
m1 = 1.0
m2 = 1.0
# Spring constants
k1 = 0.4
k2 = 1.808
# Natural lengths
L1 = 0.0
L2 = 0.0
# Friction coefficients
b1 = 0.0
b2 = 0.0

#Non linear coefficients
mu1=-1/6
mu2=-1/10

# Initial conditions
# x1 and x2 are the initial displacements; y1 and y2 are the initial velocities
x1 = -0.6
y1 = 1/2
x2 = 3.001
y2 = 5.9

# ODE solver parameters
abserr = 1.0e-8
relerr = 1.0e-6
stoptime = 200
numpoints = 2000

# Create the time samples for the output of the ODE solver.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
p = [m1, m2, k1, k2, L1, L2, b1, b2,mu1,mu2]
w0 = [x1, y1, x2, y2]

# Call the ODE solver.
wsol = odeint(vectorfield, w0, t, args=(p,),
              atol=abserr, rtol=relerr)

with open('nonlinearity3.dat', 'w') as f:
    # Print & save the solution.
    for t1, w1 in zip(t, wsol):
        print  ( t1, w1[0], w1[1], w1[2], w1[3], file=f )
```

```python
%matplotlib inline

t, x1, xy, x2, y2 = loadtxt('nonlinearity3.dat', unpack=True)
lw=1

figure(1, figsize=(6, 4.5))
plot(x1,xy, 'c', linewidth=lw)
title('Phase plot for x1')
grid(True)
savefig('ejem3.3_plot1.png', dpi=100)
```

```python
figure(2, figsize=(6, 4.5))
plot(x2,xy, 'b', linewidth=lw)
title('Phase plot for x2')
grid(True)
savefig('ejem3.3_plot2.png', dpi=100)
```
